```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset

athletes = pd.read_csv('G:/kaggle/olympics/athlete_events.csv')
regions = pd.read_csv('G:/kaggle/olympics/noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#join both datasets

athletes_df = athletes.merge(regions,how='left',on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#make col name consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True)
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
#describe func

athletes_df.describe()
```

```python
#null values check
nan_values = athletes_df.isna()
# nan_values
nan_columns = nan_values.any()
nan_columns
```

```python
#team Pakistan
athletes_df.query('Team == "Pakistan"').head()
```

```python
#top countries 
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#plot top 10 countries

plt.figure(figsize=(12,6))
plt.title("Top 10 Countries")
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
```

```python
#age distribution of the athletes

plt.figure(figsize=(12,6))
plt.title("Age Distribution of Athletes")
plt.xlabel("Age")
plt.ylabel("No of athletes")
plt.hist(athletes_df.Age, bins=np.arange(10,80,2),color='orange',edgecolor='white')
```

```python
winter_sports = athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
summer_sports = athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
#gender wise participation
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#pie chart for gender wise participation

plt.figure(figsize=(12,6))
plt.pie(gender_counts,labels=gender_counts.index, 
        autopct='%1.1f%%', startangle=180, shadow=True)
```

```python
# Total no of Medals
athletes_df.Medal.value_counts()
```

```python
#total no of female in each olympics

female_athletes = athletes_df[(athletes_df.Sex=='F')
                             &(athletes_df.Season=='Summer')][['Sex','Year']]
female_athletes = female_athletes.groupby('Year').count().reset_index()
female_athletes
```

```python
womenOlypics = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')]
womenOlypics
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenOlypics,palette='Spectral')
plt.title("Women Olympics")
```

```python
part = womenOlypics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title("Female Athletes over time")
```

```python
#gold Medal 
goldmedals = athletes_df[(athletes_df.Medal=='Gold')]
goldmedals.head()
```

```python
#exclude NaN
goldmedals = goldmedals[np.isfinite(goldmedals['Age'])]

```

```python
#goldmedal beyond age 60
goldmedals['ID'][goldmedals['Age']>60].count()
```

```python
#sporting events in which gold is won after age 60
sporting_event = goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

```python
#plot for sporting event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold after age of 60")
```

```python
#gold from each country
# goldmedals
goldmedals.Region.value_counts().reset_index(name='Medal').head(10)
```

```python
total_goldmedals = goldmedals.Region.value_counts().reset_index(name='Medal').head(10)
g=sns.catplot(x='index',y='Medal',data=total_goldmedals,
             height=5,kind='bar',palette='rocket')
# g.despine(left=True)
g.set_xlabels("Top 10 countries")
g.set_ylabels("No of medals")
plt.title("Gold Medals per Country")
plt.xticks(rotation=45)
# g.set_xticklabels(g.get_xticklabels(),rotation=45)
```
